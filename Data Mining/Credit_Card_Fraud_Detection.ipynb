```python
#data = data.fillna(data.mean(), inplace = True)
data = data.dropna()
data
```

```python
data.replace(to_replace = ['PAYMENT', 'TRANSFER', 'CASH_OUT', 'DEBIT', 'CASH_IN'],
                    value = [1, 2, 3, 4, 5], inplace = True)
data
```

```python
y = data["isFraud"]
y
```

```python
X = data[['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
print("Accuracy of classification: ", model.score(X_test, y_test))
```
