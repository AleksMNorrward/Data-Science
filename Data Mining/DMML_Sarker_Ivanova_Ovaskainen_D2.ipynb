# Preprocessing

```python
# Dropping the index column and checking the shape before and after
print("\n\033[1mData Shape Before Dropping Index:\033[0m", data.shape)
data_cleaned_unnamed = data.drop(columns=['Unnamed: 0'])
print("\n\033[1mData Shape After Dropping Index:\033[0m", data_cleaned_unnamed.shape)
```

```python
# Optimize Memory Usage
def optimize_dataframe(df):
    # Convert data types to reduce memory footprint
    for col in df.columns:
        col_type = df[col].dtype
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
    return df

# Optimize DataFrame before heavy computations
data_cleaned_unnamed = optimize_dataframe(data_cleaned_unnamed)
```

```python
# Central configuration for key variables
key_variables_config = {
    'key_variables': ['ArrDelay', 'DepDelay', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'UniqueCarrier', 'Dest', 'Origin', 'Month', 'DayOfWeek'],
    'all_variables': data_cleaned_unnamed.columns.tolist()
}
```

```python
# Clean missing values based on configuration
def clean_missing_values(df, key_vars):
    print("\n\033[1mInitial Missing Values per Column:\033[0m")
    display(df.isna().sum())

    # Remove rows where any of the specified key variables have missing values
    cleaned_df = df.dropna(subset=key_vars)

    # Check the remaining missing values in the dataset
    print("\n\033[1mMissing Values per Column After Cleaning:\033[0m")
    display(cleaned_df.isna().sum())

    # Display the shape of the data before and after cleaning
    print("\n\033[1mData Shape Before Cleaning:\033[0m", df.shape)
    print("\033[1mData Shape After Cleaning:\033[0m", cleaned_df.shape)

    return cleaned_df
```

```python
# Function to remove outliers based on configuration
def remove_outliers(df, key_vars):
    print("\n\033[1mData Shape Before Removing Outliers:\033[0m", df.shape)

    # Process only numeric columns
    numeric_vars = df[key_vars].select_dtypes(include=[np.number]).columns.tolist()

    for column in numeric_vars:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filtering the data without outliers
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    print("\n\033[1mData Shape After Removing Outliers:\033[0m", df.shape)
    return df

```

```python
# Function to select features based on configuration
def feature_selection(df, key_vars):
    print("\n\033[1mFeatures Before Selection:\033[0m", df.columns.tolist())

    # Keeping only selected features
    df_selected = df[key_vars] if all(item in df.columns for item in key_vars) else df

    print("\n\033[1mFeatures After Selection:\033[0m", df_selected.columns.tolist())
    return df_selected
```

```python
# Transform continuous numeric features into categorical (binary) features

def transform_features(df, key_vars):
    print("\n\033[1mData Shape Before Transformation:\033[0m", df.shape)

    # Filter to ensure only numeric columns are considered for binarization
    numeric_vars = df.select_dtypes(include=[np.number]).columns.tolist()

    for var in numeric_vars:
        if var in key_vars:
            # Using `.loc` to modify columns directly
            df.loc[:, f'{var}_Binarized'] = np.where(df[var] > 15, 1, 0)

    print("\n\033[1mData Shape After Transformation:\033[0m", df.shape)
    return df
```

```python
# Using the centralized key variables for operations
data_cleaned_key_NA = clean_missing_values(data_cleaned_unnamed, key_variables_config['key_variables'])
data_no_outliers_key = remove_outliers(data_cleaned_key_NA, key_variables_config['key_variables'])
data_selected_features_key = feature_selection(data_no_outliers_key, key_variables_config['key_variables'])
data_transformed_key = transform_features(data_selected_features_key, key_variables_config['key_variables'])
```

```python
# Function to one-hot encode categorical variables using get_dummies
def encode_with_get_dummies(df):
    print("\n\033[1mData Shape Before Encoding:\033[0m", df.shape)

    # Using `get_dummies` to one-hot encode all categorical columns
    df_encoded = pd.get_dummies(df, drop_first=True)

    print("\n\033[1mData Shape After Encoding:\033[0m", df_encoded.shape)
    return df_encoded
```

```python
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

```python
# Convert to DataFrame Without IPCA
data_reduced_df = pd.DataFrame(data_encoded_key)
```

```python
# Binary Conversion Function
def convert_to_boolean(df, threshold=0.5):
    return df.applymap(lambda x: True if x > threshold else False)
```

```python
# Apply the binary conversion to the reduced data
binary_data_key = convert_to_boolean(data_reduced_df)
```

```python
# Set parameters for the Apriori algorithm
min_support = 0.03
min_confidence = 0.6

# Run the Apriori algorithm to find frequent itemsets
frequent_itemsets_apriori = apriori(binary_data_key, min_support=min_support, use_colnames=True)

# Print frequent itemsets found by Apriori
print("\n\033[1mFrequent Itemsets by Apriori:\033[0m")
print(frequent_itemsets_apriori)
```

```python
# Generate association rules from the frequent itemsets
rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=min_confidence)

# Print association rules derived from the frequent itemsets
print("\n\033[1mAssociation Rules from Apriori:\033[0m")
print(rules_apriori)
```

```python
# Set parameters for the FP-Growth algorithm
min_support_fp = 0.03
min_confidence_fp = 0.6

# Run the FP-Growth algorithm to find frequent itemsets
frequent_itemsets_fp_growth = fpgrowth(binary_data_key, min_support=min_support_fp, use_colnames=True)

# Print frequent itemsets found by FP-Growth
print("\n\033[1mFrequent Itemsets by FP-Growth:\033[0m")
print(frequent_itemsets_fp_growth)

# Generate association rules from the frequent itemsets
rules_fp_growth = association_rules(frequent_itemsets_fp_growth, metric="confidence", min_threshold=min_confidence_fp)

# Print association rules derived from the frequent itemsets
print("\n\033[1mAssociation Rules from FP-Growth:\033[0m")
print(rules_fp_growth)
```
