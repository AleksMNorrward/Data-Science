## **Non-numerical Features Exploration**

### **Cancellation Code**

```python
data["CancellationCode"]
```

```python
data['CancellationCode'].unique()
```

In the context of delayed flights, "CancellationCode" refers to a code that indicates the reason for the cancellation of a flight. These codes are typically provided by airlines or aviation authorities to categorize the various reasons for flight cancellations. The codes may vary between different airlines or aviation regulatory bodies, but they usually follow a standardized format.

Some common cancellation codes include:

**A**: Carrier.

**B**: Weather.

**C**: National Air System.

**D**: Security.

**N**: No cancellation.


These codes help to identify the primary reason for the cancellation, which can be useful for analysis and tracking purposes in the aviation industry. For instance, weather-related cancellations might be more prevalent in certain regions or during specific seasons, while cancellations due to carrier issues might indicate operational challenges within an airline.
